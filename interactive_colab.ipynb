### 6. Plot each of the taxi and socio variables at the same time interactively.

```python
m = widgets.interactive(plot_taxi_socio_interactive,
                        taxi_col=taxi_cols,
                        socio_col=socio_cols,
                        top_n=range(5, 80, 5))
display(m)
```

#### Findings:
- Not a lot of taxi trips happened where the proportion of disabled population was higher, namely in many community districts in Bronx. It might be because it's inconvenient for disabled people to travel by taxi; subway ridership in Bronx to be checked.
- Median household income was higher where most taxi rides happened, namely in many community districts in Manhattan. It might be because people with higher income prefer to take taxis since the population in Manhattan is not significantly higher than in other boroughs. Conversely, poverty rate was lower and unemplyment rate was higher where most taxi rides happened; subway ridership in Manhattan to be checked.
- Mean travel time to work was lower where most taxi trips happened and they were strongly correlated in opposite directions (correlation coefficient: -0.7258912). Having said that, it's unclear whether more taxi rides contribute to shorter mean travel time to work.
- Population densities seem to be higher in Manhattan and Bronx; but the number of taxi trips was a lot higher in Manhattan than Bronx.
- Serious crime rate was highest in Midtown-Flatiron-Union Square (MN05) where the total number of taxi rides was also highest. It might be because people prefer to take taxi due to a very high serious crime rate in that community district; subway ridership in the community district to be checked.
- Minute per mile based on dropoff locations was shorter where car-free commute rate was higher. It might be that fewer cars contributed to shorter minute per mile. Also, mean travel time to work was shorter in those locations, which could be due to a similar reason.
- Average congestion surcharge was higher where mean travel time to work was longer, with a correlation coefficient of -0.9140945, which is a very strong negative correlation. It would be reasonable to believe that congestion caused the longer mean travel time to work. It is interesting that car-free commute rate is also highly correlated with average congestion surcharge (correlation coefficeint 0.6885972), which is somewhat counterintuitive. It might be because it is a rate, not a definite number of cars on the road, meaning there might still be a lot of cars on the road where car-free commute rate is higher. Or it might be because people choose not to drive a car to get to work due to frequent congestion.
* Definition of car-free commute: The percentage of workers who commute primarily by foot, bicycle, or public transportation, as a share of all workers over the age of 16 who do not work at home.
