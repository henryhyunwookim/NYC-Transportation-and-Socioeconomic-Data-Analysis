```python

def download_by_name(fname, output=None, quiet=False):
    if fname:
        url = 'https://furmancenter.org/files/NDP/'+ fname
        gdown.download(url, output=output, quiet=quiet)
    else:
        print('The file', fname, 'can not be found!')
```

```python
# Download Brooklyn, Bronx,Queens, Staten Islan 
download = 1
bk, bx, mn, qn, si =[],[],[],[],[]

if download == 1:     
    for i in range(1,19):
        nn = str(i).zfill(2)
        download_by_name('BK' + nn + '_NeighborhoodDataProfile.xlsx')
        bk.append('BK' + nn)
    for i in range(1,13):
        nn = str(i).zfill(2)          
        download_by_name('BX' + str(i).zfill(2) + '_NeighborhoodDataProfile.xlsx')
        bx.append('BX' + nn)
    for i in range(1,13):
        nn = str(i).zfill(2)
        download_by_name('MN' + nn + '_NeighborhoodDataProfile.xlsx')
        mn.append('MN' + nn)
    for i in range(1,15):
        nn = str(i).zfill(2)
        download_by_name('QN' + nn + '_NeighborhoodDataProfile.xlsx')
        qn.append('QN' + nn)
    for i in range(1,4):
        nn = str(i).zfill(2)
        download_by_name('SI' + nn + '_NeighborhoodDataProfile.xlsx')
        si.append('SI' + nn)
        
    

```

```python
bx
```

```python
#add also CD code
cds = ["BK",'BX','MN','QN','SI']
cd_all = [bk,bx,mn,qn,si]
code_all = [100,200,300,400,500]
j=0
df_cds = None
code = None
for cd in cds:
    i = 1
#     print(cd)
    for name in cd_all[j]:
        s_name = cd + ' ' + str(i).zfill(2) + ' Data'
        code = code_all[j] + i 
#         print(code)
#         print(s_name)
        if df_cds is None:
            df_cds = pd.read_excel(open(name + '_NeighborhoodDataProfile.xlsx', 'rb'),
                  sheet_name=s_name) 
            df_cds['CD_code'] =code
        else:
            temp = pd.read_excel(open(name + '_NeighborhoodDataProfile.xlsx', 'rb'),
                  sheet_name=s_name)
            temp['CD_code'] = code
            df_cds = df_cds.append(temp)
        
        i = i+1
    j = j+1
```
